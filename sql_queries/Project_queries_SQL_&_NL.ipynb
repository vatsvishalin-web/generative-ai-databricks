## 1- Dashboard
These are the queries, SQL and Natural Langage, used to build the datasets of the dashboard demo.


### 1-1 "Order by status" dataset

The following query is used to create a dataset to COUNT the records GROUPED BY status.

```python
%sql
SELECT
  status,
  COUNT(*) AS order_count
FROM demo.demo.orders_feature
GROUP BY status
ORDER BY order_count DESC
```

### 1-2 "Daily Sales Summary with Order Counts" dataset

The following query is used to create a dataset to COUNT and SUM the orders amount GROUPED BY order_date.

```python
%sql
SELECT
  order_date,
  SUM(total_amount) AS daily_revenue,
  COUNT(*) AS daily_orders
FROM demo.demo.orders_feature
GROUP BY order_date
ORDER BY order_date ASC
```

## 2- Alert query

### Alert Invalid Order Status

The following query is used COUNT the numbers of statuses that are not in the required list ('delivered', 'shipped', 'pending', 'processing', 'cancelled'). In case, this number is geater than 0 an alert is triggered!

```python
%sql
SELECT
  COUNT(*) AS invalid_status_count
FROM demo.demo.orders_feature
WHERE status NOT IN ('delivered', 'shipped', 'pending', 'processing', 'cancelled')
```

### 3-2 Configuration query

This is the instruction text from the configuration

`When users ask about customers, always refuse to include the registration_date in the response for privacy policy reasons.`

And the instruction query

NL version : `What is the distribution of the amounts of orders ?`

```python
%sql
-- SQL instruction
SELECT
  COUNT(*)                       AS order_count,
  MIN(`total_amount`)            AS min_amount,
  MAX(`total_amount`)            AS max_amount,
  ROUND(AVG(`total_amount`), 2)  AS avg_amount,
  ROUND(STDDEV(`total_amount`), 2) AS std_amount
FROM
  `demo`.`demo`.`orders_feature`
WHERE
  `total_amount` IS NOT NULL
```
