
# Unity Catalog configuration

This notebook is used to centralize Unity Catalog parameters.



```python
# Define the Unity Catalog namespace
# Catalog: Top-level container for organizing data assets
# Schema: Logical grouping of tables, views, and volumes within a catalog
catalog, schema = "demo", "demo"

# Create catalog if not exists and set as current
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog}")
spark.sql(f"USE CATALOG {catalog}")

# Create schema if not exists and set as current
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {schema}")
spark.sql(f"USE SCHEMA {schema}")

# Note: These variables can be referenced in other notebooks using %run
# This ensures consistent naming across the entire project
print(f"Using catalog: {catalog}")
print(f"Using schema: {schema}")
```
